```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
```

```python
# Parámetros
mu = 0           # Media para ambas distribuciones
sigma = 1        # Desviación estándar para la distribución gaussiana
b = 1            # Parámetro de escala para la distribución de Laplace
n_samples = 1000 # Número de muestras

# datos para las distribuciones
gaussian_data = np.random.normal(mu, sigma, n_samples)
laplace_data = np.random.laplace(mu, b, n_samples)

gaussian_kurtosis = kurtosis(gaussian_data)
laplace_kurtosis = kurtosis(laplace_data)

# Crear los histogramas
plt.figure(figsize=(10, 6))

# Distribución Gaussiana
plt.hist(gaussian_data, bins=30, density=True, alpha=0.6, color='blue', label='Gaussiana')

# Distribución de Laplace
plt.hist(laplace_data, bins=30, density=True, alpha=0.6, color='orange', label='Laplace')

# Graficar las funciones de densidad teóricas
x = np.linspace(-5, 5, 1000)

# Densidad de probabilidad Gaussiana
gaussian_pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
plt.plot(x, gaussian_pdf, color='blue', lw=2, label='densidad Gaussiana')

# Densidad de probabilidad Laplace
laplace_pdf = (1 / (2 * b)) * np.exp(-abs(x - mu) / b)
plt.plot(x, laplace_pdf, color='orange', lw=2, label='densidad Laplace')

# Configuración del gráfico
plt.title("Distribuciones Gaussiana y Laplace")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid()
plt.show()
```

```python
print(f"Curtosis de la distribución Gaussiana: {gaussian_kurtosis:.4f}")
print(f"Curtosis de la distribución de Laplace: {laplace_kurtosis:.4f}")
```
